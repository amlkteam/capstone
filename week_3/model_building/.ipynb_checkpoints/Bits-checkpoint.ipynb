```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Import the dataset
df = pd.read_csv('./Teacher_salary_csv.csv')
```

```python
df.head()
```

```python
# Filling NaN's with mean of the columns
df = df.fillna(df.mean())
```

```python
# IQR method
average_feedback= df['Average feedback']
average_feedback
```

```python
#Outliers using Z-score method
average_feedback_outliers=[]

    
threshold=3
average_feedback_mean = np.mean(average_feedback)
average_feedback_std =np.std(average_feedback)


for y in average_feedback:
    z_score= (y - average_feedback_mean)/average_feedback_std 
    if np.abs(z_score) > threshold:
        average_feedback_outliers.append(y)
```

```python
average_feedback_outliers
```

```python
# Outliers using IQR method
#sort the data
average_feedback = sorted(average_feedback)
quarantile1, quarantile3= np.percentile(average_feedback,[25,75])
range_quarantile = quarantile3-quarantile1
lower_bound = quarantile1 -(1.5 * range_quarantile) 
upper_bound = quarantile3 +(1.5 * range_quarantile) 

for data in average_feedback:
    if data < lower_bound or data > upper_bound:
        print("outlier is :" , data)
```

### Hence no outliers for this column


```python
# Working with categorical data
df = pd.read_csv('./Teacher_salary_csv.csv')
```

```python
# Working with categorical data
```

```python
# Taking care of the Missing data

df = df.fillna(0)
df
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
all_columns_list = list(df.columns)
all_columns_list
```

```python
# Feature Scaling
numeric_vars = [
 'years of teachingexperience',
 'Number of research publications',
 'Projects completed',
 'Ph.D students guided',
 'Average feedback',
 'Yearly Salary']
categorical_vars = [ 'Highest Degree']
preprocessor = ColumnTransformer(transformers=[
    ('Stdscalar', StandardScaler(), numeric_vars),
    ('ohe', OneHotEncoder(drop='first'), categorical_vars)])
```

```python
selected_df = df[['Highest Degree','years of teachingexperience', 'Number of research publications','Projects completed','Ph.D students guided', 'Average feedback','Yearly Salary']]
selected_df.head()
```

```python
# Splitting into training and testing
X_train, X_test, y_train, y_test = train_test_split(selected_df[['Highest Degree','years of teachingexperience', 'Number of research publications','Projects completed','Ph.D students guided', 'Average feedback','Yearly Salary']], 
                                                    selected_df[['Yearly Salary']], 
                                                    test_size=0.20, 
                                                    random_state=0)
```

```python
X_train.shape
X_test.shape
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
# Training the Multiple Linear Regression model on the Training set
regr = LinearRegression()
regr.fit(X_train_processed, y_train)

```

```python
# Predicting the Test set results
y_pred = regr.predict(X_test_processed)
```

```python
print('Training accuracy is', regr.score(X_train_processed, y_train))
print('Test accuracy is ', regr.score(X_test_processed, y_test))
```
