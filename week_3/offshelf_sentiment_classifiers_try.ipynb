```python
from textblob import TextBlob
text = 'Job openings in the cannabis industry have tripled in the past year and are expected to spike even more as edibles become legal'
blob = TextBlob(text)
for sentence in blob.sentences:
    print(sentence.sentiment)
    print(sentence.sentiment.polarity)
```

```python
def blob_predict(text):
    blob = TextBlob(text)
    for sentence in blob.sentences:
        #print(sentence.sentiment)
        #print(sentence.sentiment.polarity)
        return sentence.sentiment
```

```python
sentences = [headline, hl2, hl3, hl4]

for s in sentences:
    print(blob_predict(s))
```

```python
### try another sentiment analyser
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def check_compound_score(sentence):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(sentence)
    return scores['compound']
```

```python
sentences = [headline, hl2, hl3, hl4]

for s in sentences:
    print(s)
    print(check_compound_score(s))
```

```python
## textblob custom classifier

from textblob.classifiers import NaiveBayesClassifier
train = [('I love this sandwich.', '1'),
   ('This is an amazing place!', '1'),
     ('I feel very good about these beers.', '2'),
    ('I do not like this restaurant', '0'),
     ('I am tired of this stuff.', '-1'),
   ("I can't deal with this", '0'),
   ("My boss is horrible.", "-1") ]

cl = NaiveBayesClassifier(train)

cl.classify("I feel amazing!")
#'pos'

blob = TextBlob("The beer is good. But the hangover is horrible.", classifier=cl)
for s in blob.sentences:
    print(s)
    print(s.classify())
```

```python
## only count confident predictions

confident_pred = 0
correct = 0
for tup in to_train_list[3877:]:
    print(tup[0])
    gold_label = tup[1]
    prob_dist = cl.prob_classify(tup[0])
    pred_label = prob_dist.max()
    if prob_dist.prob(pred_label) > 0.9:
        confident_pred += 1
        print(gold_label, pred_label)
        if pred_label == gold_label:
            correct += 1
```
